```python
from Bio import SeqIO
```

```python
total_len = 0
count = 0
max_len = 0
min_len = 10000000000000
```

```python
for record in SeqIO.parse("all.fas", "fasta"):

    if len(record) > max_len:
        max_len = len(record)

    if len(record) < min_len:
        min_len = len(record)

    count += 1
    total_len += len(record)
```

```python
norm = lambda x: round(x, 2)

avg_len = total_len / count

avg_len = norm(avg_len)
min_len = norm(min_len)
max_len = norm(max_len)
```
